```python
try:
    with pdfplumber.open("Weber's Protestant Ethic.pdf") as pdf:
        second_page = pdf.pages[1]
        text = second_page.extract_text()
        print("Text:", text)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Process the text
doc = nlp(text)
```

```python
# Iterate over tokens and print POS tags
for token in doc:
    print(token.text, token.pos_)
```

```python
text = "Apple is a company based in Cupertino, California, founded by Steve Jobs."
# Process the text
doc = nlp(text)
```

```python
# Iterate over named entities and print entity text and label
for entity in doc.ents:
    print(entity.text, entity.label_)
```
